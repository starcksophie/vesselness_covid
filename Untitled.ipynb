```python
def print_hist(jsons, champ):
    div = 0
    for i in range(2, len(champ)):
        if len(champ) % i == 0:
            div = i
            continue
    print(div)
    if (div != 0):
        fig, axes = plt.subplots(len(champ) // div , div, figsize = (20,20))
        for i in range(len(champ) // div):
            for j in range(div):
                axes[i, j].hist([k[champ[div*i + j]] for k in jsons])
                axes[i, j].set_title(champ[div*i + j])
    else:
        fig, axes = plt.subplots(len(champ), 1, figsize = (10,15))
        for i in range(len(champ)):
            axes[i].hist([k[champ[i]] for k in jsons])
            axes[i].set_title(champ[i])
    return
```

```python
champs = ["component_count", "voxel_nbr", "lung_pixels", "mean_mean_all_vessel", "std_deviation", "max_max_all_vessel", "mean_max_all_vessel"]
lol = print_hist(jsons, champs)
```

```python
def print_hist_multiple_value(jsons, champ):
    fig, axes = plt.subplots(len(jsons), 1, figsize = (10,25))
    for i in range(len(jsons)):
        axes[i].set_title('patient ' + str(i))
        axes[i].hist(jsons[i][champ])
```

```python
print_hist_multiple_value(jsons, 'mean_all_vessel')
```

```python
print_hist_multiple_value(jsons, "max_all_vessel")
```
